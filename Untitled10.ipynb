```python
import numpy as np
actions = [] 
conditions = [] 
indexes=[]
indexrem=[]
totalActions = input('How many actions you want to give?') 
for i in range(0,int(totalActions)): 
    inp=input('Enter action '+str(i+1)+' '); 
    actions.append(inp)
for k in range(0,len(actions)):
    indexes.append([])
totalConditions = input('How many conditions you want to give') 
for i in range(0,int(totalConditions)): 
    x = input('Enter condition '+str(i+1)+' ') 
    conditions.append(x)
n=2**(len(conditions))
condition = np.zeros(((len(conditions)),n),dtype ='int32')
action = np.zeros(((len(actions)),n),dtype ='int32')
action=action.astype(str)
action[:,:]='-'
for i in range(len(conditions)):
    for j in range(n-1,-1,-1):
        condition[i][j] = j/(2**i)%2
condition=condition[::-1]
for i in range(n):
    for j in range(3):
        if condition[j][i]==0:
            if i not in indexes[0]:
                indexes[0].append(i)
            else:
                pass
for i in range(len(indexes[0]),n):
    if np.array_equal(condition[3:,i],[0,0,0]):
        if i not in indexes[1]:
            indexes[1].append(i)
    if np.array_equal(condition[3:,i],[1,1,1]):
        if i not in indexes[3]:
            indexes[3].append(i)
    if np.array_equal(condition[3:,i],[0,0,1]) or np.array_equal(condition[3:,i],[0,1,0]) or np.array_equal(condition[3:,i],[1,0,0]):
        if i not in indexes[2]:
            indexes[2].append(i)
    if np.array_equal(condition[3:,i],[0,1,1]) or np.array_equal(condition[3:,i],[1,0,1]) or np.array_equal(condition[3:,i],[1,1,0]):
        if i not in indexes[4]:
            indexes[4].append(i)
    else:
        pass
for i in range(len(actions)):
    for index in indexes[i]:
        action[i][index]='x'
for i in range(len(conditions)):
    print(str(conditions[i])+" : ",end='')
    for j in range(n):
        print(str(condition[i][j])+" ",end='')
    print()
for i in range(len(actions)):
    print(str(actions[i])+" : ",end='')
    for j in range(n):
        print(str(action[i][j])+" ",end='')
    print()
z=2
index3=-1
while z<len(indexes[0]):
    z=z*2
    k=0
    index3=index3+1
    x=(len(condition[0,:])-8)/2
    indexrem.append([])
    for j in range(int(x)):
        for i in range(len(conditions)):
            if condition[i][k]!=condition[i][k+1]:
                condition[i][k]=5
        indexrem[index3].append(k+1)
        if j==0 and index3==4:
            break
        k=k+2
    condition=np.delete(condition,indexrem[index3],1) 
    action=np.delete(action,indexrem[index3],1) 
    print("step "+str(index3+1))
    for i in range(len(conditions)):
        print(str(conditions[i])+" : ",end='')
        for j in range(len(condition[0,:])):
            print(str(condition[i][j])+" ",end='')
        print()
    for i in range(len(actions)):
        print(str(actions[i])+" : ",end='')
        for j in range(len(action[0,:])):
            print(str(action[i][j])+" ",end='')
        print()

```
